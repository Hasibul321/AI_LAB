## Assignment 1: To complete this task I have to write a python program to display values of y when 'x' is an independent variable having values [-10, 10, step_size = 0.1]

### importing necessary modules 

```python
import numpy as np 
import matplotlib.pyplot as plt
```

### declearing constants

```python
w = 4
b = 9
```

### 1: y = wx + b

```python
def linear_function(x,w,b):
    return w*x + b

x_val = np.arange(-10,10,0.1)
y_val = linear_function(x_val,w,b)

plt.figure(figsize=(12, 6))


plt.plot(x_val,y_val,label = f'y = {w}x + {b}')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("1 : Linear function")

plt.legend()

plt.grid(True)
plt.axhline(0, color='black',linewidth=2)
plt.axvline(0, color='black',linewidth=2)
plt.show()
```

### 2: y = x^2

```python
def square_function(x):
    return x*x

x_val = np.arange(-10,10,0.1)
y_val = square_function(x_val)

plt.figure(figsize=(12, 6))


plt.plot(x_val,y_val,label = f'y =  x^2')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("2 : Square function")

plt.legend()

plt.grid(True)
plt.axhline(0, color='black',linewidth=2)
plt.axvline(0, color='black',linewidth=2)
plt.tight_layout()
plt.show()

```

### 3 : y = 1/ (1 + e^(-x)); where ^ means 'to the power'

```python
def function3(x):
    return 1 / (1 + np.exp(-x))

x_val = np.arange(-10,10,0.1)
y_val = function3(x_val)

plt.figure(figsize=(12, 6))


plt.plot(x_val,y_val,label = f'y = 1/ (1 + e^(-x))')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("3 : y = 1/ (1 + e^(-x))")

plt.legend()

plt.grid(True)
plt.axhline(0, color='black',linewidth= 0.5)
plt.axvline(0, color='black',linewidth= 0.5)
plt.tight_layout()
plt.show()
```

### y = (e^x - e^(-x)) / (e^x - e^(-x))

```python
def function4(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) - np.exp(-x))

x_val = np.arange(-10,10,0.1)
y_val = function4(x_val)

plt.figure(figsize=(12, 6))


plt.plot(x_val,y_val,label = f'y = (e^x - e^(-x)) / (e^x - e^(-x))')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("4 : y = (e^x - e^(-x)) / (e^x - e^(-x))")

plt.legend()

plt.grid()
plt.tight_layout()
plt.show()
```

### 5 : y = g(f(x)) where u = f(x) = wx + b and y = g(u) = 1/ (1 + e^(-u))

```python
def function5(x,w,b):
    u = w*x + b
    return 1/(1+ np.exp(-u))

x_val = np.arange(-10,10,0.1)
y_val = function5(x_val,w,b)

plt.figure(figsize=(12,6))

plt.plot (x_val,y_val,label = f'y = g(f(x))')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("5 : y = g(f(x))")

plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

### 6 : y = g(f(x)) where f(x) = wx + b, u = f(x), and g(u) = (e^u - e^(-u)) / (e^u + e^(-u))

```python
def function6(x,w,b):
    u = w*x + b
    return (np.exp(u) - np.exp(-u)) / (np.exp(u) + np.exp(-u))

x_val = np.arange(-10,10,0.1)
y_val = function6(x_val,w,b)

plt.figure(figsize=(13,6))

plt.plot(x_val,y_val,label = f'y = g(f(x))')

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("6 : y = f(g(x))")

plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

### 7 : y1 = g1(f1(x)) where f1(x) = w1x + b1, u1 = f1(x), and g1(u1) = 1/ (1 + e^(-u1))
           y2 = g2(f2(x)) where f2(x) = w2x + b2, u2 = f2(x), and g2(u2) = 1/ (1 + e^(-u2))
           y = g3(f3(v)) where f3(v) = w3y1 + w4y2 + b, w = f3(v), and g3(w) = 1/ (1 + e^(-w))

```python
w1 = -3
b1 = -5
w2 = 3
b2 = 5
w3 = 4
w4 = 7
```

```python
def function7(x,w1,b1,w2,b2,w3,w4,b):
    u1 = w1*x + b1
    y1 = 1 / (1 + np.exp(-u1))

    u2 = w2*x + b2
    y2 = 1 /(1+ np.exp(-u2))

    w5 = w3*y1 + w4*y2 + b

    return 1 / (1 + np.exp(-w5))

x_val = np.arange(-10,10,0.1,dtype=np.float64)
y_val = function7(x_val,w1,b1,w2,b2,w3,w4,b)


plt.figure(figsize=(13,10))
plt.plot(x_val,y_val,label = f'y = g3(f3(v))')

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("7 : y = g3(f3(v))")

plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
