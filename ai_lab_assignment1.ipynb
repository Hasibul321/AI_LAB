```python
import numpy as np 
import matplotlib.pyplot as plt
```

### declearing constants

```python
a = 5
b = 9
c = -3
```

### For linear function : y = ax + b

```python
def linear_function(x,a,b):
    return a*x + b

x_val = np.linspace(-20,20,400)
y_val = linear_function(x_val,a,b)

plt.figure(figsize=(12, 6))


plt.plot(x_val,y_val,label = f'y = {a}x + {b}')

plt.xlabel('x')
plt.ylabel('y')
plt.title("Linear function")

plt.legend()

plt.grid(True)
plt.axhline(0, color='black',linewidth=2)
plt.axvline(0, color='black',linewidth=2)
plt.show()
```

### For duadratic equation  y = ax^2 +bx+c

```python
def quardratic_function(x,a,b,c):
    return a*x**2 + b*x + c

x_val = np.linspace(-20,20,400)
y_val = quardratic_function(x_val,a,b,c)

plt.figure(figsize=(12, 6))


plt.plot(x_val,y_val,label = f'y = {a}x^2 + {b}x {c}')

plt.xlabel('x')
plt.ylabel('y')
plt.title("Quadratic function")

plt.legend()

plt.grid(True)
plt.axhline(0, color='black',linewidth=2)
plt.axvline(0, color='black',linewidth=2)
plt.tight_layout()
plt.show()

```

### for sin function y = a sin(x) + b

```python
a1 = 2
b1 = 1
```

```python
def sin_function(x,a,b):
    return a * np.sin(x) + b

x_val = np.linspace(-10,10,400)
y_val = sin_function(x_val,a1,b1)

plt.figure(figsize=(12, 6))


plt.plot(x_val,y_val,label = f'y = {a1}sin (x) + {b1}')

plt.xlabel('x')
plt.ylabel('y')
plt.title("Sine function")

plt.legend()

plt.grid(True)
plt.axhline(0, color='black',linewidth=2)
plt.axvline(0, color='black',linewidth=2)
plt.tight_layout()
plt.show()
```

```python
a2 = 7
b2 = 5
```

### for exponential function y = ae^x + b

```python
def exponential_function(x, a, b):
    return a * np.exp(x) + b

x_val = np.linspace(-10,10,400)
y_val =exponential_function(x_val,a2,b2)

plt.figure(figsize=(12, 6))


plt.plot(x_val,y_val,label = f'y = {a2}e^x + {b2}')

plt.xlabel('x')
plt.ylabel('y')
plt.title("Exponential function")

plt.legend()

plt.grid()
plt.axhline(0, color='black',linewidth=2)
plt.axvline(0, color='black',linewidth=2)
plt.tight_layout()
plt.show()
```
